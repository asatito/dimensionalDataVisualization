### The proof of the pudding...

... is in visualizing it! The code below should produce a scatter plot. Don't worry about the code itself; that's what we're going to discuss during this training.

````python
import pandas as pd
from sklearn.datasets import load_iris
from umap import UMAP
from plotnine import *

iris = load_iris()
umap = UMAP()

df = pd.DataFrame(umap.fit_transform(iris.data), columns = ["x", "y"])
df["target"] = iris.target_names[iris.target]

ggplot(df, aes("x", "y", color="target")) + geom_point() + ggtitle("Iris dataset visualized using UMAP")
````
