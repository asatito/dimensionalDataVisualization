````python
df = data.faithfuld.copy()
tsne = TSNE(perplexity=30)
X = tsne.fit_transform(df)
df["x"] = X[:,0]
df["y"] = X[:,1]
ggplot(df, aes("x", "y", color="waiting")) + geom_point()
````

````python
p = 30

df = data.faithfuld.copy()
df -= df.min()
df /= df.abs().max()
tsne = TSNE(perplexity=p)
X = tsne.fit_transform(df)
df["x"] = X[:,0]
df["y"] = X[:,1]
ggplot(df, aes("x", "y", color="waiting")) + geom_point()
````

````python
p = 5

df = data.faithfuld.copy()
df -= df.min()
df /= df.abs().max()
tsne = TSNE(perplexity=p)
X = tsne.fit_transform(df)
df["x"] = X[:,0]
df["y"] = X[:,1]
ggplot(df, aes("x", "y", color="waiting")) + geom_point()
````
